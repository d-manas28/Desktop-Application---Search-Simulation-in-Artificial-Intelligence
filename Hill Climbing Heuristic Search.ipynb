```python
def is_grid_valid(input_grid):
  count_inversion = 0;
  grid_length = 9;
  if input_grid[-1] == 'B':
    grid_length = 8;
    for i in range(7):
      for j in range(i+1,8):
        if input_grid[i] == 'B' or input_grid[j] == 'B':
          continue;
        if (int(input_grid[i]) > int(input_grid[j])):
          count_inversion = count_inversion+1;
  return ((count_inversion%2) == 0);
```

```python
def get_blank_pos(start_grid):
  return start_grid.index('B')
```

```python
def calculate_manDistance(temp_grid):
  sum = 0
  # Complete this function
  for i in temp_grid:
    if i == 'B':
      continue
    counter=0
    destined_place=(int)(i)
    current_place=temp_grid.index(i)
    current_place+=1
    while(current_place!=destined_place):
      if(current_place<destined_place):
        if(destined_place - (current_place+3) >= 0):
          current_place+=3
          counter+=1
        else:
          current_place+=1
          counter+=1
      elif(current_place>destined_place):
        if((current_place - 3)- destined_place>=0):
          current_place-=3
          counter+=1
        else:
          current_place-=1
          counter+=1
    sum=sum+counter
  return sum


# calculate_manDistance(start_grid)
```

```python
def take_move(current_grid_format,blank_pos,move_type,heu_type):
  current_grid = current_grid_format[0]
  current_steps = current_grid_format[1]
  current_heuristic = current_grid_format[2]
  grid_temp_copy = copy.deepcopy(current_grid)
  # Logic of Swapping:
  # Possible swaps are: 
  # UP-> Can only be done if the B is in 2 or 3 row, i.e index = 3 to 8 (>2) 
  # Down -> Can only be done if the B is in 1 or 2 row i.e index = 0 to 5 (<6)
  # Left -> Can only be done if the B is in 2 or 3 col  i.e index = no_of_col % 3 > 0 
  # Right -> Can only be done if the B is in 1 or 2 col i.e index = no_of_col % 3 < 2

  if move_type == 'D' and blank_pos < 6:
    grid_temp_copy[blank_pos],grid_temp_copy[blank_pos+3] = grid_temp_copy[blank_pos+3],grid_temp_copy[blank_pos];
    current_steps = current_steps+1;
    if(heu_type == 1):
      current_heuristic = calculate_misTile(grid_temp_copy)  
    else:
      current_heuristic = calculate_manDistance(grid_temp_copy)

  elif move_type == 'L' and (blank_pos%3) > 0:
    grid_temp_copy[blank_pos],grid_temp_copy[blank_pos-1] = grid_temp_copy[blank_pos-1],grid_temp_copy[blank_pos];
    current_steps = current_steps+1;
    if(heu_type == 1):
      current_heuristic = calculate_misTile(grid_temp_copy)  
    else:
      current_heuristic = calculate_manDistance(grid_temp_copy)
    

  elif move_type == 'U' and blank_pos > 2:
    grid_temp_copy[blank_pos],grid_temp_copy[blank_pos-3] = grid_temp_copy[blank_pos-3],grid_temp_copy[blank_pos];
    current_steps = current_steps+1;
    if(heu_type == 1):
      current_heuristic = calculate_misTile(grid_temp_copy)  
    else:
      current_heuristic = calculate_manDistance(grid_temp_copy)

  
  elif move_type == 'R' and (blank_pos%3) < 2:
    grid_temp_copy[blank_pos],grid_temp_copy[blank_pos+1] = grid_temp_copy[blank_pos+1],grid_temp_copy[blank_pos];
    current_steps = current_steps+1;
    if(heu_type == 1):
      current_heuristic = calculate_misTile(grid_temp_copy)  
    else:
      current_heuristic = calculate_manDistance(grid_temp_copy)


  # current_steps = current_steps+1;
  return [grid_temp_copy,current_steps,current_heuristic];  
```

```python
 def calculate_misTile(temp_grid):
  counter = 0
  for i in range(1,10):
    if(temp_grid[i-1] == 'B'):
      continue
    elif(i != (int)(temp_grid[i-1])):
      counter = counter+1
  return counter  
```

```python
import copy
import time


def hill_climb_mt(current_grid,blank_pos):
  target_path = []
  grid_copy = copy.deepcopy(current_grid)
  heuristic = calculate_misTile(grid_copy)
  grid_copy_list = [grid_copy,0,heuristic]
  state_presev_list = []
  state_presev_list.append(grid_copy_list)
  target_path.append(grid_copy_list)
  new_heuristic = heuristic
  algo_start_timestamp = time.time()
  while True:  
    currently_traversing_format = state_presev_list.pop(0)
    parent_heuristic = currently_traversing_format[2] 
    for i in ['U','D','L','R']:
      updated_grid = take_move(currently_traversing_format,blank_pos,i,1)
      if(updated_grid[0] == target_grid):
        exe_time = time.time()-algo_start_timestamp
        print("Target Grid Found!")
        target_path.append(updated_grid)
        return target_path,exe_time
      elif(updated_grid[2]<parent_heuristic):
        new_heuristic = updated_grid[2]
        new_parent = updated_grid[0]
        new_state_count = updated_grid[1]
        target_path.append(updated_grid)
      
    if(new_heuristic == parent_heuristic):
      exe_time = time.time()-algo_start_timestamp
      print("\nXX !Local Maxima Found, No New Path Possible! XX\n")
      return target_path,exe_time
    else:
      state_presev_list.append([new_parent,new_state_count,new_heuristic])
```

```python
def hill_climb_manDistance(current_grid,blank_pos):
  target_path = []
  grid_copy = copy.deepcopy(current_grid)
  heuristic = calculate_manDistance(grid_copy)
  grid_copy_list = [grid_copy,0,heuristic]
  state_presev_list = []
  state_presev_list.append(grid_copy_list)
  target_path.append(grid_copy_list)
  new_heuristic = heuristic
  algo_start_timestamp = time.time()
  while True:  
    currently_traversing_format = state_presev_list.pop(0)
    parent_heuristic = currently_traversing_format[2] 
    for i in ['U','D','L','R']:
      updated_grid = take_move(currently_traversing_format,blank_pos,i,2)
      if(updated_grid[0] == target_grid):
        exe_time = time.time()-algo_start_timestamp
        print("Target Grid Found!")
        target_path.append(updated_grid)
        return target_path,exe_time
      elif(updated_grid[2]<parent_heuristic):
        new_heuristic = updated_grid[2]
        new_parent = updated_grid[0]
        new_state_count = updated_grid[1]
      
    if(new_heuristic == parent_heuristic):
      exe_time = time.time()-algo_start_timestamp
      print("\nXX !Local Maxima Found, No New Path Possible! XX\n")
      return target_path,exe_time
    else:
      state_presev_list.append([new_parent,new_state_count,new_heuristic])  
      target_path.append([new_parent,new_state_count,new_heuristic])
```
